```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter

df = pd.read_csv("/content/drive/MyDrive/Bollywood assignment/stdperformance/StudentsPerformance (2).csv")


# 1. Comparative Visualization

avg_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()

plt.figure(figsize=(8,6))
avg_scores.plot(kind="bar", figsize=(8,6), colormap="viridis")
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(title="Subjects")
plt.show()

# 2. Word Cloud
edu_counts = Counter(df["parental level of education"].dropna())

wordcloud = WordCloud(width=800, height=400,
                      background_color="white",
                      colormap="plasma").generate_from_frequencies(edu_counts)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Parental Level of Education")
plt.show()
```
